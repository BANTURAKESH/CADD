```python
import pandas as pd

# Assuming 'vd' is your DataFrame
l_columns_to_drop = ['LogP']

vd = vd.drop(columns=l_columns_to_drop)
```

```python
# split the data into x and y
x = vd.drop("vd",axis=1)

y = vd["vd"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=5)
```

```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import mean_squared_error

 
# Define the Ridge regression model
model = Ridge()

# Define a range of alpha values to search over
alphas = [0.1, 1.0, 10.0]

# Create a parameter grid for GridSearchCV
param_grid = {'alpha': alphas}

# Perform k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=5)

# Use GridSearchCV to find the best alpha using cross-validation
grid_search = GridSearchCV(model, param_grid, cv=kf, scoring='neg_mean_squared_error')
grid_search.fit(x, y)

# Get the best alpha from the grid search
best_alpha = grid_search.best_params_['alpha']

# Train the final Ridge regression model with the best alpha
final_model = Ridge(alpha=best_alpha)
final_model.fit(x, y)

# Make predictions
y_pred = final_model.predict(x)

# Evaluate the model using mean squared error
mse = mean_squared_error(y, y_pred)
print(f"Best alpha: {best_alpha}")
print(f"Mean Squared Error: {mse}")
```

```python
final_model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor

# Create a GradientBoostingRegressor with your desired hyperparameters
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3)

# Perform cross-validation
scores = cross_val_score(model, x, y, cv=5, scoring='neg_mean_squared_error')
# X is your feature matrix, and y is your target variable
# cv is the number of cross-validation folds (5 in this example)

# Calculate the mean and standard deviation of the cross-validation scores
mean_score = -scores.mean()  # Taking the negative of the mean squared error
std_score = scores.std()

# Print the results
print(f"Mean Squared Error: {mean_score}")
print(f"Standard Deviation: {std_score}")
```
